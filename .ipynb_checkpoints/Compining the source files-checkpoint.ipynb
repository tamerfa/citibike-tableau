```python
import pandas as pd
from calendar import month_abbr
```

```python
jc1801df = pd.read_csv("Resources\JC-201801-citibike-tripdata.csv")
jc1801df
```

```python
# To create dataframes for the rest of 2018 CSV files, we can use two for loops to create a list of dataframes

jc18_list = [jc1801df]
for i in range (2,10):
    jc18_list.append(pd.read_csv(f"Resources\JC-20180{i}-citibike-tripdata.csv"))
    
for i in range (10,13):
    jc18_list.append(pd.read_csv(f"Resources\JC-2018{i}-citibike-tripdata.csv"))
    
jc18_list[11].head()
```

```python
# Concatenating all 2018 dataframes into one df, with addition of a column for the month names

months = [month_abbr[i] for i in range(1,13)]
jc18df = pd.concat(jc18_list, keys=months)
jc18df = jc18df.reset_index()
del jc18df["level_1"]
jc18df = jc18df.rename(columns={"level_0":"Months"})
jc18df.tail()
```

```python
# Repeating the process for reading 2019 CSV files

jc19_list = []
for i in range (1,10):
    jc19_list.append(pd.read_csv(f"Resources\JC-20190{i}-citibike-tripdata.csv"))
    
for i in range (10,13):
    jc19_list.append(pd.read_csv(f"Resources\JC-2019{i}-citibike-tripdata.csv"))
    
jc19_list[11].head()
```

```python
# Concatenating all 2019 dataframes into one df, with addition of a column for the month names

months = [month_abbr[i] for i in range(1,13)]
jc19df = pd.concat(jc19_list, keys=months)
jc19df = jc19df.reset_index()
del jc19df["level_1"]
jc19df = jc19df.rename(columns={"level_0":"Months"})
jc19df.head()
```

```python
# Repeating the process for reading 2020 CSV files

jc20_list = []
for i in range (1,10):
    jc20_list.append(pd.read_csv(f"Resources\JC-20200{i}-citibike-tripdata.csv"))
    
for i in range (10,13):
    jc20_list.append(pd.read_csv(f"Resources\JC-2020{i}-citibike-tripdata.csv"))
    
jc20_list[11].head()
```

```python
# Concatenating all 2020 dataframes into one df, with addition of a column for the month names

months = [month_abbr[i] for i in range(1,13)]
jc20df = pd.concat(jc20_list, keys=months)
jc20df = jc20df.reset_index()
del jc20df["level_1"]
jc20df = jc20df.rename(columns={"level_0":"Months"})
jc20df.columns
```

```python
# Repeating the process for reading 2021 CSV files

jc21_list = []
for i in range (1,10):
    jc21_list.append(pd.read_csv(f"Resources\JC-20210{i}-citibike-tripdata.csv"))
    
for i in range (10,12):
    jc21_list.append(pd.read_csv(f"Resources\JC-2021{i}-citibike-tripdata.csv"))
    
jc21_list[0].usertype.unique()
```

```python
# After checking 2021 dataframes, dataframes of February till November of that year have a different columns arrangement than all the previous years
# Concatenating those dataframes together first before changing their format to match with the previous dataframes
months = [month_abbr[i] for i in range(2,12)]
jc21df = pd.concat(jc21_list[1:12], keys=months)
jc21df = jc21df.reset_index()
del jc21df["level_1"]
jc21df = jc21df.rename(columns={"level_0":"Months"})
jc21df.head()
```

```python
# Renaming the columns to match the previous years

jc21df = jc21df.rename(columns = {"started_at": "starttime",
                                 "ended_at": "stoptime",
                                 "start_station_id": "start station id",
                                 "end_station_id": "end station id",
                                 "start_station_name": "start station name",
                                 "end_station_name": "end station name",
                                 "start_lat": "start station latitude",
                                 "start_lng": "start station longitude",
                                 "end_lat": "end station latitude",
                                 "end_lng": "end station longitude",
                                 "member_casual": "usertype"})
# Replacing the values in 'usertype' to match the previous years

jc21df["usertype"] = jc21df["usertype"].replace({"member":"Subscriber", "casual":"Customer"})

# Deleting 'ride_id' and 'rideable_type' columns as they are not existent in the previous years
jc21df = jc21df.drop(columns =["ride_id", "rideable_type"], axis = 1)

# Creating 'tripduration' column for 2021 year dataframe

jc21df["starttime"] = pd.to_datetime(jc21df["starttime"])
jc21df["stoptime"] = pd.to_datetime(jc21df["stoptime"])
jc21df["tripduration"] = (jc21df["stoptime"] - jc21df["starttime"]).astype('timedelta64[s]')

# Splitting 'starttime' and 'stoptime' columns into separate columns for date and time

jc21df["start_date"] = pd.to_datetime(jc21df["starttime"]).dt.date
jc21df["start_time"] = pd.to_datetime(jc21df["starttime"]).dt.time
jc21df["stop_date"] = pd.to_datetime(jc21df["stoptime"]).dt.date
jc21df["stop_time"] = pd.to_datetime(jc21df["stoptime"]).dt.time
del jc21df["starttime"]
del jc21df["stoptime"]

# Rearranging columns

jc21df = jc21df[['Months', 'tripduration', 'start_date', 'start_time', 'stop_date', 'stop_time', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude','usertype']]
jc21df.head()
```

```python
# First, Concatenating only the years before 2021 (As they contained gender and birth year data)
p_frames = [jc18df, jc19df, jc20df]
jc18_20df = pd.concat(p_frames, keys=("2018", "2019", "2020"))
jc18_20df = jc18_20df.reset_index()
```

```python
# del jc18_20df["level_1"]
jc18_20df = jc18_20df.rename(columns={"level_0":"Years"})
jc18_20df["gender"] = jc18_20df["gender"].replace({0: "Unknown", 1:"Male", 2:"Female"})
jc18_20df.head()
```

```python
# Splitting 'starttime' and 'stoptime' columns into separate columns for date and time

jc18_20df["start_date"] = pd.to_datetime(jc18_20df["starttime"]).dt.date
jc18_20df["start_time"] = pd.to_datetime(jc18_20df["starttime"]).dt.time
jc18_20df["stop_date"] = pd.to_datetime(jc18_20df["stoptime"]).dt.date
jc18_20df["stop_time"] = pd.to_datetime(jc18_20df["stoptime"]).dt.time
del jc18_20df["starttime"]
del jc18_20df["stoptime"]
```

```python
# Rearranging columns
jc18_20df = jc18_20df[['Years', 'Months', 'tripduration', 'start_date', 'start_time', 'stop_date', 'stop_time', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude','usertype', 'birth year', 'gender']]

jc18_20df
```

```python
# Second removing bikeid, birth year and gender data columns to concatenate with 2021 data
jc_prev_df = jc18_20df.drop(["birth year","gender"], axis=1)

# Add the column 'years' to jc21df and adding the value 2021
jc21df["Years"] = "2021"
jc21df = jc21df[['Years', 'Months', 'tripduration', 'start_date', 'start_time', 'stop_date', 'stop_time', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude','usertype']]

# Concatenating all years dataframes
jctotaldf = pd.concat([jc_prev_df, jc21df])
jctotaldf.tail()
```

```python
# Download both 'jc18_20df' and 'jctotaldf' as CSV files

jc18_20df.to_csv("Resources\Cleaned_data\jc18_20.csv", index=False)
jctotaldf.to_csv("Resources\Cleaned_data\jc18_21.csv",index=False)
```
